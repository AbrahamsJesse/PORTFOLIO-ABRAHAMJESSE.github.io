```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from IPython.display import display

movie_genres = {"mystery" : "1",
                "crime": "2",
                "true crime" : "2",
                "animation" : "3",
                "cartoon" : "3",
                "cartoons" : "3",
                "drama" : "4",
                "western" : "6",
                "comedy" : "7",
                "tv movie" : "8",
                "family" : "9",
                "family friendly" : "9",
                "action" : "10",
                "documentary" : "11",
                "romance" : "12",
                "fantasy" : "13",
                "horror" : "14",
                "music" : "15",
                "musical" : "15",
                "thriller" : "16",
                "war" : "17",
                "war movie" : "17",
                "adventure" : "18",
                "history" : "19",
                "historical" : "19",
                "sci-fi" : "31",
                "science fiction" : "31",
                "fantasy" : "31",
                "biography" : "37",
                "biopic" : "37",
                "bio-pic" : "37"}

def genre_function ():
    
    def pick_genre(answer):
            while True:
                user_input = input('''
                Please enter one of the following genres available:
                
                Mystery, Crime, True crime, Animation, Cartoon, Cartoons,
                Drama, Western, Comedy, Tv movie, Family,Family Friendly,
                Action, Documentary, Romance, Fantasy, Horror, Music,
                Musical, Thriller, War, War movie, Adventure, History,
                Historical, Sci-fi, Science fiction, Fantasy, Biography,
                Biopic, Bio-pic''')
                
                if user_input in answer:
                    return answer[user_input]
                else:
                    print("Error. Please check your spelling or choose a different genre")
    genre = pick_genre(movie_genres)
    print(genre)
        
           
    n_pages = int(input("From 1 to 10, How long do you want your catalogue to be?"))
    
  
    year = int(input("Select year of release"))
    def year_func(answer):
        if answer >= 2019:
            return answer
        else:
            answer = str(answer)
            return "older-2019"
    year_func(year)


    df = pd.DataFrame()


    for i in range(1, n_pages + 1):

        url = f"https://www.fmovies.hn/filter?type=movie&quality=all&release_year={year}&genre={genre}&country=all&page={i}"
        headers = {
            "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36"
        }

        r = requests.get(url, headers=headers)
        r.status_code
        soup = BeautifulSoup(r.content)
        page = soup.find_all("div", attrs={"class": "flw-item"})

        film_dict = {}
        film_index = 0

        for item in page:
            name = item.find_all("h2", attrs={"class": "film-name"})[0].get_text().strip()
            year = item.find_all("span", attrs={"class": "fdi-item"})[0].get_text().strip()
            #duration = item.find_all("span", attrs={"class": "fdi-item fdi-duration"})[0].get_text().strip()
            #quality = item.find_all("div", attrs={"class": "pick film-poster-quality"})[0].get_text().strip()

            film_dict[film_index] = {
                "name": name,
                "year": year,
                "duration": duration,
                'quality': quality
            }

            film_index += 1
            movies = pd.DataFrame.from_dict(film_dict, orient="index")
            df = pd.concat([df,movies])

            df.reset_index(drop=True, inplace=True)
        display(df)
        return df
```

```python
def main():
    action = input('Do you want to search by genre or by actor? (Type "genre" or "actor"): ').lower()

    if action == 'genre':
        genre_function()
    elif action == 'actor':
        actor_function()
    else:
        print('Invalid input. Please enter "genre" or "actor".')

if __name__ == '__main__':
    main()
```
